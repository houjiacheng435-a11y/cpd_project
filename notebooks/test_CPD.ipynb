# CPD Example Notebook

This notebook demonstrates how to import the `cpd` package and run change point detection on the sample dataset.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from cpd import Brk

data_path = '../data/TEST.parquet(1).gzip'
df = pd.read_parquet(data_path)
print('Data shape:', df.shape)
```

```python
series = df['p']
if not isinstance(series.index, pd.DatetimeIndex):
    if 'time' in df.columns:
        series.index = pd.to_datetime(df['time'])
    else:
        raise ValueError('Series must use a DatetimeIndex or contain a time column.')

brk = Brk(method='cusum', q=1.0)
change_points = brk.detect(series)
print('Found', len(change_points), 'change points')
print(change_points)
```
